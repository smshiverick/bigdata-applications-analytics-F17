```python
opioid_evr = pd.crosstab(df['HEROINEVR'], df['PRLMISEVR'])
opioid_evr
```

```python
opioid_pct = opioid_evr.div(opioid_evr.sum(1), axis=0)
opioid_pct
```

```python
opioid_pct.plot.bar()
plt.savefig('opioids.png', bbox_inches='tight')
```

### Plot Prescription Opioid use by AgeCat
